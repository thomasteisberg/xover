```python
import icepyx as ipx
import geopandas as gpd
import earthaccess
from shapely.ops import orient
import h5py
import matplotlib.pyplot as plt
from concurrent.futures import ThreadPoolExecutor, as_completed
import threading
```

```python
def atl06_grab_threadsafe(granule, buffer):
    file_obj = earthaccess.open(granules=[granule])[0]
    atl_file = h5py.File(name=file_obj, mode="r")
    
    meta_data = dict(granule.items())
    date = gpd.pd.to_datetime(meta_data['meta']['native-id'].split('_')[1])
    
    orients = atl_file["orbit_info"]["sc_orient"][:]
    strong_beams = ["gt1l","gt2l","gt3l"] if orients == 0 else ["gt3r","gt2r","gt1r"]
    
    gdf_list = []
    for beam in strong_beams:
        gdf = gpd.GeoDataFrame(
            data={
                "h_li": atl_file[f"{beam}/land_ice_segments/h_li"][:],
                "atl06_quality_summary": atl_file[f"{beam}/land_ice_segments/atl06_quality_summary"][:],
                "beam": beam,
                "date": date,
            },
            geometry=gpd.points_from_xy(
                atl_file[f"{beam}/land_ice_segments/longitude"][:],
                atl_file[f"{beam}/land_ice_segments/latitude"][:],
            ),
            crs="4326",
        )
        gdf_list.append(gdf[gdf.within(buffer)])
    
    return gpd.pd.concat(gdf_list)
```

```python
icesat_xover = gpd.read_parquet('../data/icesat-radar-merged_points.parquet')
icesat_xover_3031_buffer = icesat_xover.buffer(10)
icesat_xover_reproj_4326= icesat_xover_3031_buffer.to_crs(4326)
```

```python
%%time
# parallelize over (granule, buffer) pairs
tasks = []
for buffer in icesat_xover_reproj_4326:
    buffer = orient(buffer, sign=1.0)
    spatial_extent = list(buffer.exterior.coords)
    granules = earthaccess.search_data(
        short_name="ATL06", cloud_hosted=True,
        polygon=spatial_extent, temporal=("2020-01-01", "2021-01-01"),
    )
    for granule in granules:
        tasks.append((granule, buffer))
```

```python
%%time
results = []
with ThreadPoolExecutor(max_workers=12) as executor:
    futures = {executor.submit(atl06_grab_threadsafe, g, b): (g, b) for g, b in tasks}
    for future in as_completed(futures):
        try:
            results.append(future.result())
        except Exception as e:
            print(f"failed: {e}")

gdf_all = gpd.pd.concat(results, ignore_index=True)
```
